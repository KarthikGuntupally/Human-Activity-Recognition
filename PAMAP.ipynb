```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Path to the directories where the PAMAP2 dataset files are stored
protocol_directory = '/content/drive/MyDrive/PAMAP2_Dataset/Protocol'
optional_directory = '/content/drive/MyDrive/PAMAP2_Dataset/Optional'

# Function to load all .dat files from a directory
def load_dat_files(directory):
    data_files = [f for f in os.listdir(directory) if f.endswith('.dat')]
    dataframes = []
    for file in data_files:
        file_path = os.path.join(directory, file)
        df = pd.read_csv(file_path, delim_whitespace=True, header=None)  # Adjust delimiter if necessary
        dataframes.append(df)
    return dataframes

# Load data from both directories
protocol_dataframes = load_dat_files(protocol_directory)
optional_dataframes = load_dat_files(optional_directory)

# Combine all DataFrames into a single DataFrame
all_data = pd.concat(protocol_dataframes + optional_dataframes, ignore_index=True)

# Preprocessing Data
# Handling missing values (you might need to adjust this based on your dataset)
all_data.dropna(inplace=True)

# Assume the last column is your target variable and the rest are features
X = all_data.iloc[:, :-1]  # Features
y = all_data.iloc[:, -1]   # Target variable

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert target variable to binary form using a threshold (e.g., 0.5)
threshold = 0.5
y_binary = (y > threshold).astype(int)

# Splitting the Dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_binary, test_size=0.2, random_state=42)

# Logistic Regression
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(X_train, y_train)

# Predictions on the testing set
y_pred = logistic_regression_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
